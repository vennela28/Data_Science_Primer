# PyPlot

- Matplotlib is one of the important plotting libraries in Python
- There are many tools. But <b>matplotlib.pyplot</b> is a collection of command style functions that make matplotlibwork like MATLAB.

- Each pyplot function makes some changes to the figure.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 4, 6, 4])
plt.ylabel('Numbers')
plt.xlabel('Indicies')
plt.title("MyPlot")
plt.show()
```

<h4> plot x versus y </h4>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid() # grid on

plt.show()
```

- For every x, y pair of arguments, there is an optional third argument which is the <b> format string </b>that indicates the color and the line type of the plot.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid() # grid on

plt.show()
```

```python
import numpy as np
t = np.arange(0, 5, 0.2)

plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

<h3> Controlling line properties </h3>

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth=5.0)
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]

lines = plt.plot(x1, y1, x2, y2)

# use keyword args
plt.setp(lines[0], color='r', linewidth=2.0)

# MATLAB style string value pairs
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)

plt.grid()
```

<h3> working with multiple figures and axes </h3>

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.figure(1)

# The subplot() command specifies numrows, numcols,
#fignum where fignum ranges from 1 to numrows*numcols.

plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1, 2, 3])
plt.subplot(212)
plt.plot([4, 5, 6])

plt.figure(2)
plt.plot([4, 5, 6])

plt.figure(1)               # figure(1), subplot(212) - CURRENT
plt.subplot(211)            # make subplot(211) in figure(1) current
plt.title('Show 1, 2, 3')   # subplot 211 title
plt.show()
```

```python
plt.figure(1)
plt.subplot(121)
plt.plot([1, 2, 3])
plt.grid()
plt.subplot(122)
plt.plot([1, 4, 9])
plt.grid()

plt.figure(2)
plt.grid()
plt.plot([1, 8, 27], 'g-')

plt.figure(3)
plt.subplot(321)
plt.plot([4, 5, 6])
plt.grid()
plt.subplot(322)
plt.plot([9, 8, 7])
plt.grid()
plt.show()
```
